```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('results.csv')
```

```python
data_data1 = data.loc[data['Dataset'] == 'wine']
argmax1 = data_data1.iloc[data_data1['F-score'].values.argmax()]
data_data2 = data.loc[data['Dataset'] == 'iris']
argmax2 = data_data2.iloc[data_data2['F-score'].values.argmax()]
data_data3 = data.loc[data['Dataset'] == 'glass']
argmax3 = data_data3.iloc[data_data3['F-score'].values.argmax()]
data_data4 = data.loc[data['Dataset'] == 'diabetes']
argmax4 = data_data4.iloc[data_data4['F-score'].values.argmax()]
data_data5 = data.loc[data['Dataset'] == 'seeds']
argmax5 = data_data5.iloc[data_data5['F-score'].values.argmax()]
```

```python
fil_data = pd.DataFrame([argmax1, argmax2, argmax3, argmax4, argmax5])
fil_data
```

```python
filtered_data = data.loc[data['CV'] == 5].loc[data['Voting'] == 'distance'].loc[data['Metric'] == 'euclidean'].loc[data['Dataset'] == 'iris']
```

```python
filtered_data
```

```python
filtered_data
```

```python
cv = filtered_data['CV']
plt.plot(cv, filtered_data['Accuracy'], label='Accuracy')
plt.plot(cv, filtered_data['Precision'], label='Precision')
plt.plot(cv, filtered_data['Recall'], label='Recall')
plt.plot(cv, filtered_data['F-score'], label='F-score')
plt.legend()
plt.xlabel('Wielkość kroswalidacji')
plt.ylabel('Wartości metryk')
plt.title('Wykres zależeności wielkości kroswalidacji od liczby sąsiadów')
plt.show()
```
